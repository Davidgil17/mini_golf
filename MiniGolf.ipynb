```python
golfers = pd.read_csv('scores.csv', names = ['Sir Barton (1)', 'Gallant Fox (2)', 'Omaha (3)', 'War Admiral (4)', 'Whirlaway (5)', 'Count Fleet (6)', 'Assault (7)', 'Citation (8)', 'Secretariat (9)', 'Seattle Slew (10)',  'Affirmed (11)', 'American Pharoah (12)', 'Justify (13)', 'Next TC Winner (14)', 'Damascus (15)', 'Alydar (16)', 'Northern Dancer (17)', 'Man O War (18)'])
```

```python
totals = []
averages = []
medians = []
for i in range (0,9):
    totals.append(golfers.iloc[i].sum())
    averages.append(golfers.iloc[i].mean().round(1))
    medians.append(golfers.iloc[i].median())
    

golfers["Total"] = totals
golfers["Mean"] = averages
golfers["Median"] = medians
```

```python
appended_golfers = golfers.append(golfers.sum().rename('Total')).sort_values(by=["Total"])
```

```python
appended_golfers = appended_golfers.append(golfers.mean().round(1).rename('Mean'))
```

```python
appended_golfers = appended_golfers.append(golfers.median().rename('Median'))
```
